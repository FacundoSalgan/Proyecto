```python
import pandas as pd
url = "https://raw.githubusercontent.com/FacundoSalgan/Proyecto/refs/heads/main/spotify_dataset.csv"
df = pd.read_csv(url)

# Mostrar las primeras filas del dataset
df.head()
```

```python
import matplotlib.pyplot as plt

# 1️⃣ Histograma: Distribución de la popularidad
plt.figure(figsize=(8,5))
plt.hist(df["Popularity"], bins=20, color="skyblue", edgecolor="black")
plt.xlabel("Popularidad")
plt.ylabel("Cantidad de canciones")
plt.title("Distribución de la Popularidad en Spotify")
plt.grid(True)
plt.show()

# 2️⃣ Scatter Plot: Relación entre Tempo y Energía
plt.figure(figsize=(8,5))
plt.scatter(df["tempo"], df["energy"], alpha=0.5, color="red")
plt.xlabel("Tempo (BPM)")
plt.ylabel("Energía")
plt.title("Relación entre Tempo y Energía en Canciones de Spotify")
plt.grid(True)
plt.show()

# 3️⃣ Gráfico de líneas: Evolución de la popularidad por año
df_grouped = df.groupby("year")["popularity"].mean()
plt.figure(figsize=(8,5))
plt.plot(df_grouped.index, df_grouped.values, marker="o", linestyle="-", color="green")
plt.xlabel("Año")
plt.ylabel("Popularidad Promedio")
plt.title("Evolución de la Popularidad de Canciones en Spotify")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/FacundoSalgan/Proyecto/refs/heads/main/pandora_dataset.csv"
df = pd.read_csv(url)

# Mostrar las primeras filas del dataset
df.head()
```

```python
print(df.columns)  # Ver los nombres de las columnas
print(df["Genre"].unique())  # Ver qué géneros hay en el dataset
```

```python
df["Genre"] = df["Genre"].str.lower()  # Convierte todo a minúsculas
```

```python
df_pop = df[df["Genre"] == "pop"]
```

```python
df_popularity = df_pop.groupby("Release Year")["Popularity"].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df_popularity["Release Year"], df_popularity["Popularity"], marker="o", linestyle="-", color="b", label="Popularidad promedio")

plt.xlabel("Año")
plt.ylabel("Popularidad promedio")
plt.title("Evolución de la Popularidad del Pop a lo largo de los años")
plt.legend()
plt.grid(True)

plt.show()
```

```python
df_genre_streams = df.groupby("Genre")["Streams/Views (Millions)"].sum().reset_index()
```

```python
df_genre_streams = df_genre_streams.sort_values(by="Streams/Views (Millions)", ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(df_genre_streams["Genre"], df_genre_streams["Streams/Views (Millions)"], color="c")

plt.xlabel("Género Musical")
plt.ylabel("Total de Streams")
plt.title("Streams Totales por Género")
plt.xticks(rotation=45)  # Rotar etiquetas para mejor visibilidad
plt.grid(axis="y", linestyle="--", alpha=0.7)  # Cuadrícula para mejor lectura

plt.show()
```

```python
df_Language = df.groupby("Language").sum().reset_index()
df_Language = df_Language.sort_values(by="Streams/Views (Millions)", ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(df_Language["Language"], df_Language["Streams/Views (Millions)"], color="c")

plt.xlabel("lenguaje")
plt.ylabel("Total de Streams")
plt.title("Streams Totales por lenguaje")
plt.xticks(rotation=45)  # Rotar etiquetas para mejor visibilidad
plt.grid(axis="y", linestyle="--", alpha=0.7)  # Cuadrícula para mejor lectura

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.histplot(df["Popularity"], bins=30, kde=True, color="blue")

plt.xlabel("Popularidad")
plt.ylabel("Frecuencia")
plt.title("Distribución de la Popularidad de las Canciones en Spotify")

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="Genre", y="Streams/Views (Millions)", data=df, palette="Set2")

plt.xlabel("Género")
plt.ylabel("Cantidad de Streams")
plt.title("Distribución de Streams por Género en Spotify")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x="Popularity", y="Streams/Views (Millions)", data=df, alpha=0.5, color="red")

plt.xlabel("Popularidad")
plt.ylabel("Streams")
plt.title("Relación entre Popularidad y Streams en Spotify")
plt.grid(True)

plt.show()
```
